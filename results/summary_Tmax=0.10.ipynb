# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Plot all columns of df_PPO_R2_p1 with respect to alpha in one plot:
def multi_plot_scatter_with_lines(df, 
    legend_labels=['Final tax rate', 'Mean tax rate', 'Median tax rate', 'Std tax rate', 'Max tax rate', 'Min tax rate'], 
    colors=['blue', 'red', 'green', 'orange', 'black', 'purple'], 
    title='Tax rates for percentile vs Alpha'):
    """
    Plots scatter plots with connecting lines for specified columns against a common x-axis.

    Parameters:
    - df: pandas DataFrame containing the data.
    - title: string, title of the plot.
    """
    # Ensure legend_labels and colors have the same length
    if len(legend_labels) != len(colors):
        raise ValueError("legend_labels and colors must have the same length.")
    
    # Sorting the DataFrame based on 'alpha' to ensure lines are plotted correctly
    df_sorted = df.sort_values(by='alpha')

    y_labels = df.columns[1:]

    # Setting up the plot
    plt.figure(figsize=(5, 3))

    # Plotting each column against 'alpha' with custom legend labels
    for y_label, legend_label, color in zip(y_labels, legend_labels, colors):
        plt.scatter(df_sorted['alpha'], df_sorted[y_label], label=legend_label, color=color, alpha=0.5)
        plt.plot(df_sorted['alpha'], df_sorted[y_label], color=color, linestyle='--', alpha=0.5)

    # Adding legend, grid, labels, and title
    plt.legend()
    plt.grid(True)
    plt.xlabel('Alpha')
    plt.ylabel('Values')
    plt.title(title)

    # Setting the xlim and ylim based on the min and max values of 'alpha' and y_labels
    x_min, x_max = df_sorted['alpha'].min(), df_sorted['alpha'].max()
    y_min = min(df_sorted[label].min() for label in y_labels)
    y_max = max(df_sorted[label].max() for label in y_labels)
    plt.xlim(x_min - 0.05, x_max + 0.05) # Slightly wider range for clarity
    plt.ylim(y_min - 0.05, y_max + 0.05) # Slightly wider range for clarity

    # Display the plot
    plt.show()
```

```python
def plot_scatter_with_lines(df, legend_labels=['T_max', 'T_esp', 'T_scale'], colors=['blue', 'red', 'green'], title='T_max, T_esp, T_scale vs Alpha'):
    """
    Plots scatter plots with connecting lines for specified columns against a common x-axis.

    Parameters:
    - df: pandas DataFrame containing the data.
    - legend_labels: list of strings, labels for the legend corresponding to the columns to plot.
    - colors: list of strings, colors for each plot.
    - title: string, title of the plot.
    """
    # Ensure legend_labels and colors have the same length
    if len(legend_labels) != len(colors):
        raise ValueError("legend_labels and colors must have the same length.")
    
    # Fixed column names
    y_labels = ['T_max', 'T_esp', 'T_scale']

    # Sorting the DataFrame based on 'alpha' to ensure lines are plotted correctly
    df_sorted = df.sort_values(by='alpha')

    # Setting up the plot
    plt.figure(figsize=(5, 3))

    # Plotting each column against 'alpha' with custom legend labels
    for y_label, legend_label, color in zip(y_labels, legend_labels, colors):
        plt.scatter(df_sorted['alpha'], df_sorted[y_label], label=legend_label, color=color, alpha=0.5)
        plt.plot(df_sorted['alpha'], df_sorted[y_label], color=color, linestyle='--', alpha=0.5)

    # Adding legend, grid, labels, and title
    plt.legend()
    plt.grid(True)
    plt.xlabel('Alpha')
    plt.ylabel('Values')
    plt.title(title)

    # Setting the xlim and ylim based on the min and max values of 'alpha' and y_labels
    x_min, x_max = df_sorted['alpha'].min(), df_sorted['alpha'].max()
    #y_min = min(df_sorted[label].min() for label in y_labels)
    #y_max = max(df_sorted[label].max() for label in y_labels)
    plt.xlim(x_min - 0.05, x_max + 0.05) # Slightly wider range for clarity
    #plt.ylim(y_min - 0.05, y_max + 0.05) # Slightly wider range for clarity

    # Display the plot
    plt.show()

```
